```python
def copyax(axp,ax):
    
    ar=20
    
    xd=axp.collections[0].get_offsets()[:,0]
    yd=axp.collections[0].get_offsets()[:,1]
    cd=axp.collections[0].get_facecolor()
    ax.scatter(xd,yd, s=ar,c=cd ,alpha=0.8)
   
    ax.set_title(axp.get_title(loc='left'),loc='left')
    ax.set_title(axp.get_title(loc='right'),loc='right')
    ax.set_xlabel(axp.get_xlabel(),**flabel)
    ax.set_ylabel(axp.get_ylabel(),**flabel)
    ax.axvline(0, color='black',lw=0.8, alpha=0.5)
    ax.axhline(0, color='black',lw=0.8, alpha=0.5)
    ax.set_xticks(axp.get_xticks(),labels=axp.get_xticklabels())
    ax.set_xlim(axp.get_xlim())
    ax.set_ylim(axp.get_ylim())

    
    zmap.changelabel(ax,**flabel)
    zmap.changebox(ax,'all',linewidth=1.5)
    return
```

```python
fig, axu = plt.subplots(2, 2, figsize=(9,8),constrained_layout=True)

copyax(ax1,axu[0,0])
copyax(ax2,axu[0,1])
copyax(ax3,axu[1,0])
copyax(ax4,axu[1,1])

labfile =  f'Fig1.pdf'
#plt.savefig(labfile)
```
